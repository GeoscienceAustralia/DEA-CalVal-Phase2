```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('RAW_DATA/FLAME_DATA/temp2')
```

```python
df = df.set_index('Wavelength')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(5.5, 5.5))
df.plot(ax=axes, legend=False)
axes.set_xlabel('Wavelength (nm)')
plt.savefig('flamespec.png')
```
