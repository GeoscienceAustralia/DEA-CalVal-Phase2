```python
def ReadInK(alldata):
    kf = pd.read_csv('/g/data/up71/projects/CalVal_Phase1/RAW_DATA/Panels/GA_10inch.txt', sep=' ', skiprows=3)
    kf.set_index('Wavelength', inplace=True)
    
    tempSer = pd.Series()
    for i in alldata.filename.unique():
        tempSer = pd.concat([tempSer, kf])
        
    return tempSer
```

```python
def ApplyKtoRadiance(alldata, kvals, radcol):
    alldata['kf'] = kvals['K-factor'].values
    alldata['rad2'] = alldata[radcol]/alldata.kf
    alldata.rename(columns={radcol: 'oldrad', 'rad2': radcol}, inplace=True)
    
    return alldata
```

```python
kvals = ReadInK(allSR3500dataM3)
allSR3500dataM3 = ApplyKtoRadiance(allSR3500dataM3, kvals, 'Rad_Target')
kvals = ReadInK(allSR3500dataM4)
allSR3500dataM4 = ApplyKtoRadiance(allSR3500dataM4, kvals, 'Rad_Target')
kvals = ReadInK(allSR3500dataM6)
allSR3500dataM6 = ApplyKtoRadiance(allSR3500dataM6, kvals, 'Rad_Target')
kvals = ReadInK(allASDdataM5a)
allASDdataM5a = ApplyKtoRadiance(allASDdataM5a, kvals, 'radiance')
kvals = ReadInK(allASDdataM5b)
allASDdataM5b = ApplyKtoRadiance(allASDdataM5b, kvals, 'radiance')
kvals = ReadInK(allASDdataM6)
allASDdataM6 = ApplyKtoRadiance(allASDdataM6, kvals, 'radiance')
```

```python
def MakeMeanSpec(allASDdata, allASDSpec, startWavelength, endWavelength):
        dftemp = pd.DataFrame([allASDdata.date_saved.unique(), allASDSpec[np.logical_and(allASDSpec.index >startWavelength, allASDSpec.index < endWavelength)].mean(axis=0)]).T
        dftemp.columns = ['date_saved', 'rad_mean']
        return dftemp
```
