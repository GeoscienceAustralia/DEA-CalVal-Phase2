```python
import numpy as np
allASDGround = allASDdata[np.logical_and(allASDdata.date_saved > pd.Timestamp(2020,7,31,2,25,45), allASDdata.date_saved < pd.Timestamp(2020,7,31,2,28,20))]
allSR3500Ground = allSR3500data[np.logical_and(allSR3500data.date_saved > pd.Timestamp(2020,7,31,2,23,30), allSR3500data.date_saved < pd.Timestamp(2020,7,31,2,26,0))]

allASDPanel = allASDdata[np.logical_or(allASDdata.date_saved > pd.Timestamp(2020,7,31,2,28,30), allASDdata.date_saved < pd.Timestamp(2020,7,31,2,25,30))]
allSR3500Panel = allSR3500data[np.logical_and(allSR3500data.date_saved > pd.Timestamp(2020,7,31,2,20), allSR3500data.date_saved < pd.Timestamp(2020,7,31,2,23,10))]
```

```python
allASDPanel = allASDPanel[~allASDPanel.Spec_number.isin([22,24,25])]
```

```python
import MakeSpecDF
```

```python
allSR3500Ground['Line'] = 1
allSR3500Panel['Line'] = 1
```

```python
allSR3500Ground.rename(columns={'Rad_Target':'radiance'}, inplace = True)
allSR3500Panel.rename(columns={'Rad_Target':'radiance'}, inplace = True)
```

```python
#allASDground.reset_index(inplace=True)
allASDGroundSpec = MakeSpecDF.make_spec_df(allASDGround)
allSR3500GroundSpec = MakeSpecDF.make_spec_df(allSR3500Ground)

allASDPanelSpec = MakeSpecDF.make_spec_df(allASDPanel)
allSR3500PanelSpec = MakeSpecDF.make_spec_df(allSR3500Panel)
```

```python
allASDGroundSpec.plot(legend=False)
allSR3500GroundSpec.plot(legend=False)
```

```python
allASDPanelSpec.plot(legend=False)
allSR3500PanelSpec.plot(legend=False)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8.5, 8.5))
plt.tight_layout(pad=1.5, w_pad=2.5, h_pad=3.5)
allASDGroundSpec.mean(axis=1).plot(ax=axes[1,0], color='b')
allSR3500GroundSpec.mean(axis=1).plot(ax=axes[1,0], color='r')
allASDGroundSpec.plot(ax=axes[0,0], legend=False)
allSR3500GroundSpec.plot(ax=axes[0,1], legend=False)
(allASDGroundSpec.mean(axis=1)/allSR3500GroundSpec.mean(axis=1)).plot(ax=axes[1,1])

axes[0,0].set_title('ASD ground spectra')
axes[0,1].set_title('SR3500 ground spectra')
axes[1,0].set_title('Averaged spectra (ASD=blue, SR3500=red)')
axes[1,1].set_title('Ratio spectrum (Mean ASD/Mean SR3500)')

for i in range(2):
    for j in range(2):
        axes[i,j].set_xlabel('Wavelength (nm)')
        if i+j!=2:
            axes[i,j].set_ylabel('Radiance (W m$^{-2}$ nm$^{-1}$ sr$^{-1}$)')

plt.savefig('fig2g.png')
```
